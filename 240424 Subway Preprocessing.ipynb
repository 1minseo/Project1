역주소 및 전화번호 데이터 병합

```python
add = pd.read_csv("./서울교통공사 역주소 및 전화번호_20230831.csv", encoding='cp949')
add.head()
```

```python
# '역명' 컬럼의 각 값을 문자열로 변환 후 뒤에 '역' 붙이기
add['역명'] = add['역명'].astype(str) + '역'
add.head()
```

```python
add = add.drop(columns=['연번','역번호'])
add.head()
```

```python
# '호선' 컬럼의 각 값을 문자열로 변환 후 뒤에 '호선' 붙이기
add['호선'] = add['호선'].astype(str) + '호선'
add.head()
```

```python
add['역명2'] = add['호선'] + ' ' + add['역명']
add.head()
```

```python
# 데이터 불러오기
df = pd.read_csv("./subway_seoul.csv")
```

```python
# '역명' 컬럼을 기준으로 두 데이터프레임을 합치기
merged_df = pd.merge(df, add, on='역명2')
merged_df.head()
```

```python
merged_df = merged_df.drop(columns=['역명_y','노선명'])
merged_df.head()
```

```python
# '역명_x' 컬럼명 변경
merged_df.rename(columns={'역명_x':'역명'}, inplace=True)
merged_df.head()
```

```python
# 공백을 기준으로 문자열 분할
merged_df['도로명주소'].str.split(" ")
```

```python
# expand=True 옵션으로 리스트를 프레임으로 변경
merged_df['도로명주소'].str.split(" ", expand=True)
```

```python
merged_df["자치구"] = merged_df["도로명주소"].str.split(" ", expand=True)[1]
merged_df.head()
```

```python
merged_df.to_csv("./subway_seoul(+address).csv", index=False)
```
