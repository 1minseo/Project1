##### 기상자료개방포털 기온 분석
- 폭염: 일 최고기온이 33°C 이상인 날
- 한파: 아침 최저기온(03:01~09:00)이 영하 12도 이하인 날

```python
tem19 = pd.read_csv("./2019 기온분석.csv", encoding='cp949')
tem20 = pd.read_csv("./2020 기온분석.csv", encoding='cp949')
tem21 = pd.read_csv("./2021 기온분석.csv", encoding='cp949')
tem22 = pd.read_csv("./2022 기온분석.csv", encoding='cp949')
tem23 = pd.read_csv("./2023 기온분석.csv", encoding='cp949')

tem19.head()
```

```python
tem = pd.concat([tem19, tem20, tem21, tem22, tem23])
tem.head()
```

```python
tem['날짜'] = tem['날짜'].str.replace('\t','')
tem.head()
```

```python
tem.to_csv("./temperature_seoul.csv", index=False)
```

##### 기상자료개방포털 미세먼지 분석

```python
dust19 = pd.read_csv("./일별평균대기오염도_2019.csv", encoding='cp949')
dust20 = pd.read_csv("./일별평균대기오염도_2020.csv", encoding='cp949')
dust21 = pd.read_csv("./일별평균대기오염도_2021.csv", encoding='cp949')
dust22 = pd.read_csv("./일별평균대기오염도_2022.csv", encoding='cp949')
dust23 = pd.read_csv("./일별평균대기오염도_2023.csv", encoding='cp949')

dust19.head()
```

```python
dust = pd.concat([dust19, dust20, dust21, dust22, dust23])
dust.head()
```

```python
dust.rename(columns={'측정일시':'날짜'}, inplace=True)
dust.head()
```

```python
# '날짜' 컬럼의 데이터를 날짜 형식으로 변환
dust['날짜'] = pd.to_datetime(dust['날짜'], format='%Y%m%d')
dust['날짜'] = dust['날짜'].dt.strftime('%Y-%m-%d')

dust.head()
```

```python
dust.to_csv("./dust_seoul.csv", index=False)
```

```python
# 기온 데이터, 강수량 데이터 합치기
tem = pd.read_csv("./temperature_seoul.csv")
pre = pd.read_csv("./precipitation_seoul.csv")
```

```python
# 공통된 '날짜' 컬럼을 기준으로 merge
weather = pd.merge(tem, pre, on="날짜")
weather.head()
```

```python
weather.drop(['지점_x', '지점_y'], axis=1, inplace=True)
weather.head()
```

```python
weather.to_csv("./pre_tem_merge.csv", index=False)
```
