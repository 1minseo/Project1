```python
df19_1 = pd.read_csv("./BUS_STATION_BOARDING_MONTH_201901.csv", encoding='cp949')
df19_2 = pd.read_csv("./BUS_STATION_BOARDING_MONTH_201902.csv", encoding='cp949')
df19_3 = pd.read_csv("./BUS_STATION_BOARDING_MONTH_201903.csv", encoding='cp949')
df19_4 = pd.read_csv("./BUS_STATION_BOARDING_MONTH_201904.csv", encoding='cp949')
df19_5 = pd.read_csv("./BUS_STATION_BOARDING_MONTH_201905.csv", encoding='cp949')
df19_6 = pd.read_csv("./BUS_STATION_BOARDING_MONTH_201906.csv", encoding='cp949')
df19_7 = pd.read_csv("./BUS_STATION_BOARDING_MONTH_201907.csv", encoding='cp949')
df19_8 = pd.read_csv("./BUS_STATION_BOARDING_MONTH_201908.csv", encoding='cp949')
df19_9 = pd.read_csv("./BUS_STATION_BOARDING_MONTH_201909.csv", encoding='cp949')
df19_10 = pd.read_csv("./BUS_STATION_BOARDING_MONTH_201910.csv", encoding='cp949')
df19_11 = pd.read_csv("./BUS_STATION_BOARDING_MONTH_201911.csv", encoding='cp949')
df19_12 = pd.read_csv("./BUS_STATION_BOARDING_MONTH_201912.csv", encoding='cp949')

bus19 = pd.concat([df19_1, df19_2, df19_3, df19_4, df19_5, df19_6, df19_7, df19_8, df19_9, df19_10, df19_11, df19_12])
bus19.head()
```

```python
# 자치구 데이터 불러오기
df_excel = pd.read_excel("./자치구정류장코드분류.xlsx", dtype=str)
df_excel.head()
```

```python
bus19['앞2자리'] = bus19['버스정류장ARS번호'].astype(str).str[:2]
bus19.head()
```

```python
df_excel['앞2자리'] = df_excel['버스정류장ARS번호'].astype(str).str[:2]
df_excel.head()
```

```python
# 두 데이터프레임을 '앞2자리' 컬럼을 기준으로 합침
merged_bus19 = pd.merge(bus19, df_excel, on='앞2자리')
merged_bus19.head()
```

```python
merged_bus19.drop(columns=['등록일자', '앞2자리', '버스정류장ARS번호_y', '정류장ID(앞3자리)'], inplace=True)
merged_bus19.head()
```

```python
# 컬럼명 변경
merged_bus19.rename(columns={'버스정류장ARS번호_x':'버스정류장ARS번호'}, inplace=True)
merged_bus19.head()
```

```python
merged_bus19.to_csv("2019 bus_seoul.csv", index=False)
```
